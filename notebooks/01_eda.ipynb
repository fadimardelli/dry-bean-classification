```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("../data/Dry_Bean_Dataset.xlsx")
df.head()
```

```python
df.info(), df.describe().T
```

```python
df['Class'].value_counts(normalize=True)
```

```python
ax = df.drop(columns=['Class']).hist(figsize=(12,10))
plt.tight_layout(); plt.show()
```

```python
corr = df.drop(columns=['Class']).corr(numeric_only=True)
plt.figure(figsize=(9,7))
plt.imshow(corr, interpolation='nearest'); plt.colorbar(); plt.title("Correlation")
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90, fontsize=8)
plt.yticks(range(len(corr.columns)), corr.columns, fontsize=8)
plt.tight_layout(); plt.show()
```
